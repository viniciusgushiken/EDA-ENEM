## 3) Quantos cursos novos abrem por ano? 


```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# data set do curso
%time df_curso = pd.read_csv("/Users/vinic/Desktop/Digital House/data/educacao_superior/microdados_ed_superior_2018/dados/DM_CURSO.CSV", encoding = 'latin1', sep="|")
```

```python
#notamos varios NaN, data não está no formato YYYY-MM-DD
df_curso_data = df_curso['DT_INICIO_FUNCIONAMENTO']
df_curso_data
```

```python
df_curso_data = df_curso_data.reset_index()
df_curso_data
```

```python
#tirando os NaN, dataype das datas está em object e nao em formato de data

df_curso_data = df_curso_data.dropna()
df_curso_data.info()
```

```python
#transformando em formato de data, paramentro dayfirst indica que o dia está em primeiro, e o corece é para ignorar valores extremos

df_curso_data['DT_INICIO_FUNCIONAMENTO']= pd.to_datetime(df_curso_data['DT_INICIO_FUNCIONAMENTO'], dayfirst=True, errors = 'coerce')
df_curso_data.info()
```

```python
#separando ano, mes e dia

df_curso_data['Year'] = df_curso_data['DT_INICIO_FUNCIONAMENTO'].dt.year
df_curso_data['Month'] = df_curso_data['DT_INICIO_FUNCIONAMENTO'].dt.month
df_curso_data['Day'] = df_curso_data['DT_INICIO_FUNCIONAMENTO'].dt.day
```

```python
df_curso_data
```

```python
ano = df_curso_data.groupby(['Year']).size() #agrupando o numero de cursos por ano
ano = ano.reset_index() # resetando o index
ano = ano.rename(columns={0:"Cursos"}) #renomeando a coluna de contagem de cursos
ano['Year'] = ano['Year'].astype(int) #mudando o tipo de dado da coluna ano

ano
```
